### **B. Áp Dụng Kiến Thức Numpy**

#### 1. Basic Numpy Operations

```python
# Chuyển đổi cột 'raisedhands' thành mảng numpy và tính các thống kê cơ bản
raisedhands_scores = student_data['raisedhands'].to_numpy()
mean_raisedhands = np.nanmean(raisedhands_scores)
median_raisedhands = np.nanmedian(raisedhands_scores)
std_raisedhands = np.nanstd(raisedhands_scores)
print("Mean Raised Hands:", mean_raisedhands)
print("Median Raised Hands:", median_raisedhands)
print("Standard Deviation of Raised Hands:", std_raisedhands)
```

##### Giải thích:

- ``to_numpy()``: Chuyển đổi cột 'raisedhands' thành mảng numpy.
- ``np.nanmean()``: Tính giá trị trung bình của mảng, bỏ qua các giá trị NaN.
- ``np.nanmedian()``: Tính giá trị trung vị của mảng, bỏ qua các giá trị NaN.
- ``np.nanstd()``: Tính độ lệch chuẩn của mảng, bỏ qua các giá trị NaN.

#### 2. Handling Missing Values

```python
# Kiểm tra các giá trị thiếu trong dataset
missing_values = student_data.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

- ``student_data.isnull().sum()``: Kiểm tra và đếm số lượng giá trị bị thiếu trong mỗi cột của DataFrame student_data.

=> Các giá trị trả về 0, nghĩa là các dữ liệu không bị thiếu hay trống ở ô nào cả.

#### 3. Data Transformation

```python
# Thêm epsilon vào các giá trị trong cột 'raisedhands' để tránh lỗi log(0)
epsilon = 1e-10
log_raisedhands = np.log(student_data['raisedhands'] + epsilon)

# Hiển thị dữ liệu đã biến đổi
print("Original Raised Hands:", student_data['raisedhands'][:5])
print("Log Transformed Raised Hands:", log_raisedhands[:5])
```

##### Giải thích:

- ``np.log()``: Áp dụng phép biến đổi log tự nhiên cho các giá trị trong mảng.
- Thêm epsilon (1e-10) vào các giá trị trong cột 'raisedhands' để tránh lỗi log(0).

#### 4. Data Aggregation

```python
# Tổng hợp 'raisedhands' theo 'gender'
gender_raisedhands = student_data.groupby('gender')['raisedhands'].sum().to_numpy()

# Hiển thị dữ liệu tổng hợp
print("Raised Hands by Gender:", gender_raisedhands)
```

##### Giải thích:

- ``groupby()``: Nhóm dữ liệu theo 'gender'.
- ``sum()``: Tính tổng số lần giơ tay cho từng nhóm gender.
- ``to_numpy()``: Chuyển đổi kết quả thành mảng numpy.

#### 5. Data Normalization

```python
# Chuẩn hóa Min-Max của 'raisedhands'
min_raisedhands = np.min(student_data['raisedhands'])
max_raisedhands = np.max(student_data['raisedhands'])
normalized_raisedhands = (student_data['raisedhands'] - min_raisedhands) / (max_raisedhands - min_raisedhands)
print("Normalized Raised Hands:", normalized_raisedhands[:5])
```

##### Giải thích:

- ``np.min()``: Tìm giá trị nhỏ nhất trong mảng.
- ``np.max()``: Tìm giá trị lớn nhất trong mảng.

### **C. Thực Hiện Làm Sạch Dữ Liệu - Data Cleaning**

```python
# Kiểm tra các giá trị bị thiếu
missing_values = student_data.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Điền các giá trị thiếu bằng giá trị trung vị của cột tương ứng
student_data['raisedhands'] = student_data['raisedhands'].fillna(student_data['raisedhands'].median())
student_data['VisITedResources'] = student_data['VisITedResources'].fillna(student_data['VisITedResources'].median())
student_data['AnnouncementsView'] = student_data['AnnouncementsView'].fillna(student_data['AnnouncementsView'].median())
student_data['Discussion'] = student_data['Discussion'].fillna(student_data['Discussion'].median())

# Xác minh các giá trị bị thiếu đã được xử lý
print("Missing values after filling:\n", student_data.isnull().sum())
```

### **D. Áp Dụng Kiến Thức Matplotlib - Data Visualization**

```python
# Thiết lập môi trường vẽ biểu đồ
%matplotlib inline
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.figsize'] = (10, 10)
matplotlib.rcParams['font.size'] = 15

# Vẽ biểu đồ histogram cho toàn bộ các cột trong dataset
student_data.hist(figsize=(15, 15), edgecolor='black')
plt.suptitle('Histogram for All Columns')
plt.show()
```

#### 1. Biểu Đồ Phân Phối Raised Hands (Raised Hands Distribution)

```python
plt.figure(figsize=(10, 6))
plt.hist(student_data['raisedhands'], bins=30, edgecolor='black')
plt.title('Phân Phối Raised Hands')
plt.xlabel('Số Lần Giơ Tay')
plt.ylabel('Tần Suất')
plt.show()
```

#### 2. Biểu Đồ Box Plot Cho VisITed Resources Theo Giới Tính (VisITedResources by Gender)

```python
plt.figure(figsize=(12, 8))
student_data.boxplot(column='VisITedResources', by='gender', grid=False)
plt.title('VisITedResources Theo Giới Tính')
plt.xlabel('Giới Tính')
plt.ylabel('VisITed Resources')
plt.show()
```

#### 3. Biểu Đồ Scatter Plot Giữa Raised Hands và Discussion (Raised Hands vs. Discussion)

```python
plt.figure(figsize=(10, 6))
plt.scatter(student_data['raisedhands'], student_data['Discussion'], alpha=0.5)
plt.title('Raised Hands vs. Discussion')
plt.xlabel('Raised Hands')
plt.ylabel('Discussion')
plt.show()
```

#### 4. Biểu Đồ Thanh Cho Raised Hands Theo Quan Hệ Với Học Sinh (Raised Hands by Relation)

```python
plt.figure(figsize=(10, 6))
relation_raisedhands = student_data.groupby('Relation')['raisedhands'].mean()
relation_raisedhands.plot(kind='bar', edgecolor='black')
plt.title('Raised Hands Theo Quan Hệ Với Học Sinh')
plt.xlabel('Quan Hệ Với Học Sinh')
plt.ylabel('Raised Hands Trung Bình')
plt.show()
```

#### 5. Biểu Đồ Tròn (Pie Chart) Cho Phụ Huynh Trả Lời Khảo Sát (ParentAnsweringSurvey)

```python
plt.figure(figsize=(5, 5))
parent_survey_counts = student_data['ParentAnsweringSurvey'].value_counts()
plt.pie(parent_survey_counts, labels=parent_survey_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Tỷ Lệ Phụ Huynh Trả Lời Khảo Sát')
plt.axis('equal')  # Đảm bảo biểu đồ tròn được vẽ dưới dạng hình tròn
plt.show()
```

#### 6. Biểu Đồ Đường (Line Plot) Cho Số Lần Xem Thông Báo Theo Học Kỳ (AnnouncementsView by Semester)

```python
plt.figure(figsize=(10, 6))
semester_announcements = student_data.groupby('Semester')['AnnouncementsView'].mean()
semester_announcements.plot(kind='line', marker='o')
plt.title('Số Lần Xem Thông Báo Theo Học Kỳ')
plt.xlabel('Học Kỳ')
plt.ylabel('Announcements View')
plt.show()
```

##### Giải thích:

- Biểu Đồ Histogram Cho Các Cột Liên Quan Đến Số Lượng: Vẽ biểu đồ histogram để hiển thị phân phối của tất cả các cột số trong dataset.
- Phân Phối Raised Hands: Vẽ biểu đồ histogram để hiểu rõ hơn về phân phối số lần giơ tay của học sinh.
- Box Plot Cho VisITed Resources Theo Giới Tính: Vẽ biểu đồ box plot để so sánh số lần truy cập tài nguyên học tập theo từng nhóm giới tính.
- Scatter Plot Giữa Raised Hands và Discussion: Vẽ biểu đồ scatter plot để thể hiện mối quan hệ giữa số lần giơ tay và tham gia thảo luận.
- Thanh Cho Raised Hands Theo Quan Hệ Với Học Sinh: Vẽ biểu đồ thanh để phân tích số lần giơ tay theo quan hệ với học sinh.
- Pie Chart Cho Phụ Huynh Trả Lời Khảo Sát: Vẽ biểu đồ tròn để hiển thị tỷ lệ phụ huynh trả lời khảo sát.
- Line Plot Cho Số Lần Xem Thông Báo Theo Học Kỳ: Vẽ biểu đồ đường để hiển thị số lần xem thông báo theo học kỳ.
